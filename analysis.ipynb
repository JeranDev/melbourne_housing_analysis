## Analysis

Recently, the housing market in Melbourne has seen explosive growth. In this analysis, we will conduct an exploratory data analysis to analyze trends in the housing market and try to determine trends that might be useful to an investor.


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("housing.csv")
```

### Data Exploration


```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Plot 1: Size of land vs. property cost by property type


```python
sns.scatterplot(x="Landsize", y="Price", hue="Type", data=df)
plt.show()
```

It looks like the "house" type increases in price dramatically as the landsize goes up.


### Plot 2: Average price of housing each year


```python
year_vs_price_df = df.groupby("YearBuilt", as_index=False).agg({"Price": "mean"})
year_vs_price_df
```

Using this new dataframe, we can now plot our chart using a line chart.


```python
sns.lineplot(x="YearBuilt", y="Price", data=year_vs_price_df)
plt.plot()
```

Interestingly, it looks like as the property becomes newer, the average price decreases.


### Plot 3: Average price per region


```python
region_vs_price_df = df.groupby("Regionname", as_index=False).agg({"Price": "mean"})
region_vs_price_df
```

We'll have to make the bar chart bigger to fit all of the text properly.


```python
plt.figure(figsize=[20, 8])
sns.barplot(x="Regionname", y="Price", data=region_vs_price_df)
plt.plot()
```

According to the bar chart above, the Southern Metropolitan region has the highest average prices.


### Plot 4: Histogram of property prices


```python
sns.histplot(x="Price", data=df)
plt.plot()
```

The price is positively scewed. This implied that most houses have lower prices, but there are few individuals with very high prices that stretch the distribution towards the right.


### Plot 5: Box plot of properties built in different regions over the years


```python
sns.boxplot(x="YearBuilt", y="Regionname", data=df)
plt.plot()
```

The Northern Metropolitan has the widest range of time in terms of properties built. It also looks like the Western Victoria region has built very few properties after the year 2000.
